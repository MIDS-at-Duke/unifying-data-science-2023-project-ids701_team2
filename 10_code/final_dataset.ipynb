### Data Preparation

Index(['state', 'Children 0-18', 'Adults 19-25', 'Adults 26-34',

       'Vehicle Theft']

Index here is the addition of Violent and Property Crime. Crime index usually refers to the crime propensity within a region and it's usually defined as crime per certain (100,000) populations. 

Since we are mostly interested in the violent crime, which is defined by [FBI](https://ucr.fbi.gov/crime-in-the-u.s/2018/crime-in-the-u.s.-2018/topic-pages/violent-crime) that include rape and sexual assault, robbery, assault and murder, we have a column indicating the total number of violent crime cases. 

```python
final_df['Violent_Crimes'] = final_df['Violent'] + final_df['Murder'] + final_df['Robbery'] + final_df['Rape'] + final_df['Aggravated Assault']
final_df['Violent_Crime_Rate'] = final_df['Violent_Crimes'] / final_df['Population']
```
