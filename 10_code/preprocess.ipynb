# Age Percentage

```python
import os
csv_folder = '../00_source_data/age_percentage'
csv_files = [os.path.join(csv_folder, file) for file in sorted(os.listdir(csv_folder)) if file.endswith('.csv')]
```

```python
combined_age_pct = pd.concat([pd.read_csv(file, header=2, skiprows=[3,55,56,57,58,59,60,61,62,63,64,65,66,67,68,69]) \
    for file in csv_files])

# drop the last two columns
combined_age_pct = combined_age_pct.iloc[:, :-2]

# create the column for year
combined_age_pct['year'] = np.repeat(np.arange(2008, 2022), 51)

# rename the state column
combined_age_pct.rename(columns={'Location': 'state'}, inplace=True)

# make sure col/row number is correct
assert combined_age_pct.shape[0] == 51*14
assert combined_age_pct.shape[1] == 8

combined_age_pct
```

# EST Homelessness

```python
csv_folder = '../00_source_data/Estimates_of_Homelessness'
csv_files = [os.path.join(csv_folder, file) for file in sorted(os.listdir(csv_folder)) if file.endswith('.csv')]
```

```python
combined_homelessness = pd.concat([pd.read_csv(file, header=2, \
    skiprows=[3,55,56,57,58,59,60,61,62,63,64,65,66,67,68,69,70,71,72,73,74,75,76,77]) \
    for file in csv_files])

# select the column we need
combined_homelessness = combined_homelessness.loc[:,['Location', 'Total People Experiencing Homelessness']]

# create the column for year
combined_homelessness['year'] = np.repeat(np.arange(2007, 2021), 51)

# rename the state column
combined_homelessness.rename(columns={'Location': 'state'}, inplace=True)

# make sure col/row number is correct
assert combined_homelessness.shape[0] == 51*14
assert combined_homelessness.shape[1] == 3

combined_homelessness
```

```python
combined_homelessness.isnull().any()
```

# GDP

```python
csv_folder = '../00_source_data/gdp'
csv_files = [os.path.join(csv_folder, file) for file in sorted(os.listdir(csv_folder)) if file.endswith('.csv')]
```

```python
combined_gdp = pd.concat([pd.read_csv(file, header=0, index_col=0) for file in csv_files], axis=1)

# get the first two strings of the column name
combined_gdp.columns = [col[:2] for col in combined_gdp.columns]
combined_gdp
```

```python
# convert the columns to rows
combined_gdp = combined_gdp.stack().reset_index()
combined_gdp.columns = ['year', 'state', 'gdp']
combined_gdp
```

```python
combined_gdp.isnull().any()
```

```python
# read state abbreviations
state_abbrev = pd.read_csv('../00_source_data/state_abbr.csv')
state_abbrev.head()
```

```python
# map the state abbreviation to the full name
combined_gdp['state'] = combined_gdp['state'].map(state_abbrev.set_index('code')['state'])

# slice the year
combined_gdp['year'] = combined_gdp['year'].str.slice(0,4).astype(int)

# make sure col/row number is correct
assert combined_gdp.shape[0] == 51*26
assert combined_gdp.shape[1] == 3

combined_gdp
```

```python
combined_gdp.isnull().any()
```

## Income Equality

Due to covid, dataset for 2020 is separately downloaded from census bereau.

```python
income_inequality_no_2020 = pd.read_csv('../00_source_data/Income inequality (Gini Coefficient).csv', header=4)

# select relevant columns and rows
income_inequality_no_2020 = income_inequality_no_2020[income_inequality_no_2020['Fips'] != 'US']
income_inequality_no_2020 = income_inequality_no_2020.loc[:, ['Location', 'TimeFrame', 'Data']]
income_inequality_no_2020.columns = ['state', 'year', 'gini']
income_inequality_no_2020.reset_index(drop=True, inplace=True)
income_inequality_no_2020
```

```python
income_inequality_2020 = pd.read_csv('../00_source_data/Income inequality (Gini Coefficient)_2020.csv', encoding='iso-8859-1')

# select even rows
income_inequality_2020 = income_inequality_2020.iloc[::2, :]
income_inequality_2020 = income_inequality_2020.iloc[:-1, :] 
income_inequality_2020['Label (Grouping)'] = income_inequality_2020['Label (Grouping)'].str.split('!').str[0]
income_inequality_2020['year'] = 2020
income_inequality_2020.columns = ['state', 'gini', 'year']
# switch gini and year
income_inequality_2020 = income_inequality_2020[['state', 'year', 'gini']]
income_inequality_2020.reset_index(drop=True, inplace=True)
income_inequality_2020
```

```python
# combine the two dataframes
income_inequality = pd.concat([income_inequality_no_2020, income_inequality_2020], axis=0)
income_inequality.sort_values(by=['state', 'year'], inplace=True)
income_inequality.reset_index(drop=True, inplace=True)
income_inequality
```

```python
income_inequality.isnull().any()
```

# Immigration

```python
nst_2000_2009 = pd.read_csv('../00_source_data/immigration/nst_est2009_alldata.csv', skiprows=[1,2,3,4,5,57], header=0)
nst_2010_2020 = pd.read_csv('../00_source_data/immigration/nst-est_2010-2020-alldata.csv', skiprows=[1,2,3,4,5,57], header=0)
```

```python
# select relevant columns
state_name = nst_2010_2020['NAME']
img_2000_2009 = nst_2000_2009.loc[:, nst_2000_2009.columns.str.startswith('INTERNATIONALMIG')]
img_2010_2020 = nst_2010_2020.loc[:, nst_2010_2020.columns.str.startswith('INTERNATIONALMIG')]

# combine the three dataframes
img_2000_2020 = pd.concat([state_name, img_2000_2009, img_2010_2020], axis=1)
img_2000_2020.rename(columns={'NAME':'state'}, inplace=True)
img_2000_2020
```

```python
# convert columns into rows
img_2000_2020 = img_2000_2020.melt(id_vars=['state'], var_name='year', value_name='international_migration')

# get the last four digits of the year
img_2000_2020['year'] = img_2000_2020['year'].str.slice(-4).astype(int)
img_2000_2020
```

# Unemployment rate

```python
ur = pd.read_csv('../00_source_data/Unemployment rate by Total.csv', header=4)
# select relevant columns and rows
ur = ur[ur['Fips'] != 'US']
ur = ur.loc[:, ['Location', 'TimeFrame', 'Data']]
ur.columns = ['state', 'year', 'unemployment_rate']
ur.reset_index(drop=True, inplace=True)
ur
```

# Merge to df

```python
# select relevant years and merge to the main dataframe
age_pct_08_20 = combined_age_pct.loc[combined_age_pct['year'].between(2008, 2020, inclusive='both'), :]
homelessness_08_20 = combined_homelessness.loc[combined_homelessness['year'].between(2008, 2020, inclusive='both'), :]
gdp_08_20 = combined_gdp.loc[combined_gdp['year'].between(2008, 2020, inclusive='both'), :]
race_pop_08_20 = race_pop_pivot.loc[race_pop_pivot['year'].between(2008, 2020, inclusive='both'), :]
income_inequality_08_20 = income_inequality.loc[income_inequality['year'].between(2008, 2020, inclusive='both'), :]
img_08_20 = img_2000_2020.loc[img_2000_2020['year'].between(2008, 2020, inclusive='both'), :]
ur_08_20 = ur.loc[ur['year'].between(2008, 2020, inclusive='both'), :]
```

```python
# create a list of dataframes
dfs = [age_pct_08_20, homelessness_08_20, gdp_08_20, race_pop_08_20, income_inequality_08_20, img_08_20, ur_08_20]

# initialize the merged dataframe with the first dataframe
merged_df = dfs[0]

# loop through the remaining dataframes and merge with the merged dataframe
for i in range(1, len(dfs)):
    merged_df = pd.merge(merged_df, dfs[i], on=['state', 'year'], how='left')
```
