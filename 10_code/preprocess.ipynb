# Age Percentage

```python
import os
csv_folder = '../00_source_data/age_percentage'
csv_files = [os.path.join(csv_folder, file) for file in os.listdir(csv_folder) if file.endswith('.csv')]
```

```python
combined_age_pct = pd.concat([pd.read_csv(file, index_col=0,header=2, skiprows=[3,55,56,57,58,59,60,61,62,63,64,65,66,67,68,69]) \
    for file in csv_files])

# drop the last two columns
combined_age_pct = combined_age_pct.iloc[:, :-2]

# create the column for year
combined_age_pct['year'] = np.repeat(np.arange(2008, 2022), 51)

# make sure col/row number is correct
assert combined_age_pct.shape[0] == 51*14
assert combined_age_pct.shape[1] == 7

combined_age_pct
```

# EST Homelessness

```python
csv_folder = '../00_source_data/Estimates_of_Homelessness'
csv_files = [os.path.join(csv_folder, file) for file in os.listdir(csv_folder) if file.endswith('.csv')]
```

```python
combined_homelessness = pd.concat([pd.read_csv(file, index_col=0,header=2, \
    skiprows=[3,55,56,57,58,59,60,61,62,63,64,65,66,67,68,69,70,71,72,73,74,75,76,77]) \
    for file in csv_files])

# select the column we need
combined_homelessness = combined_homelessness.loc[:,['Total People Experiencing Homelessness']]

# create the column for year
combined_homelessness['year'] = np.repeat(np.arange(2007, 2021), 51)

# make sure col/row number is correct
assert combined_homelessness.shape[0] == 51*14
assert combined_homelessness.shape[1] == 2

combined_homelessness
```
